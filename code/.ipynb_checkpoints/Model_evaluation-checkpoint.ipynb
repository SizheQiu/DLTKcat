## Enzyme level

```python
corr_data = []
uniq_ec = list( np.unique(data['EC']) )
for ec in uniq_ec:
    temp_data = data[data['EC']==ec]
    temp_data = temp_data.reset_index().drop(['index'],axis=1)
    if len(np.unique(temp_data['Temp_K'])) > 2:
        temp = {'EC':ec,'#entries':len(temp_data.index)}
        log10kcats = list( np.log10( temp_data['kcat'] ) )
        Ts = list(temp_data['Temp'])
        res = stats.pearsonr( Ts, log10kcats )
        temp['pearson_coeff'] = res[0]
        temp['p_value'] = res[1]
        corr_data.append(temp)
corr_result = pd.DataFrame(corr_data)
```

```python
sigcorr_result = ( corr_result[corr_result['p_value'] <= 0.05] ).reset_index().drop(['index'],axis=1)
print( sum(sigcorr_result['#entries']) )
print( len(sigcorr_result.index) )
classes = []
for i in range(len(sigcorr_result.index)):
    ec_number = (list( sigcorr_result['EC'] )[i]).strip()
    classes.append( 'EC_' + str(ec_number[0]))
sigcorr_result['Class'] = classes
```

```python
fig,axs = plt.subplots(1,2,figsize=(10,4))
sns.boxplot(data = sigcorr_result, x= 'Class', y='pearson_coeff', ax = axs[0])
sns.histplot(data= sigcorr_result, x= 'Class',bins = 10, ax = axs[1] )
plt.suptitle('Pearson correlations of temperature and kcat (p-value<0.05)')
```

## Enzyme-substrate level

```python
corr_EST = []
for ec in uniq_ec:
    temp_data = data[data['EC']==ec]
    temp_data = temp_data.reset_index().drop(['index'],axis=1)
    if len(np.unique(temp_data['Temp_K'])) <= 2:
        continue
    else:
        uniq_subs = np.unique(temp_data['smiles'])
    for sub in uniq_subs:
        temp_D = (temp_data[ temp_data['smiles']==sub ]).reset_index().drop(['index'],axis=1)
        if len(np.unique(temp_D['Temp_K'])) > 2:
            entry = { 'EC':ec, 'smiles':sub, '#entries':len(temp_D.index) }
            log10kcats = list( np.log10( temp_D['kcat'] ) )
            Ts = list(temp_D['Temp'])
            res = stats.pearsonr( Ts, log10kcats )
            entry['pearson_coeff'] = res[0]
            entry['p_value'] = res[1]
            corr_EST.append(entry)
corrEST_result = pd.DataFrame(corr_EST)
corrEST_result.head()
```

```python
sigcorrEST_result = ( corrEST_result[corrEST_result['p_value'] <= 0.05] ).reset_index().drop(['index'],axis=1)
print( sum(sigcorrEST_result['#entries']) )
print( len(sigcorrEST_result.index) )
classes = []
for i in range(len(sigcorrEST_result.index)):
    ec_number = (list( sigcorrEST_result['EC'] )[i]).strip()
    classes.append( 'EC_' + str(ec_number[0]))
sigcorrEST_result['Class'] = classes
```

```python
fig,axs = plt.subplots(1,2,figsize=(10,4))
sns.histplot(data=sigcorrEST_result, x= 'pearson_coeff',bins = 20, ax = axs[0] )
sns.histplot(data=sigcorrEST_result, x= 'Class',bins = 10, ax = axs[1] )
plt.suptitle('Pearson correlation of temperature and kcat (p-value<0.05)')
```

# Plot

```python
def get_rmse(x,y):
    rmse = sqrt(((x - y)**2).mean(axis=0))
    return round(rmse, 4)
def get_r2(x,y):
    return round( r2_score(x,y), 4)
```

```python
from scipy.stats import gaussian_kde
```

```python
def scatter_with_gaussian_kde(ax, x, y, size):
    xy = np.vstack([x, y])
    z = gaussian_kde(xy)(xy)
    ax.scatter(x, y, c=z, s=size)
```

```python
fig, axs = plt.subplots(1,4,figsize=(20,4))
scatter_with_gaussian_kde(axs[0], np.log10(predPCE['kcat']), predPCE['pred_log10kcat'], size = 30)
scatter_with_gaussian_kde(axs[1], np.log10(predPAFN['kcat']), predPAFN['pred_log10kcat'], size = 30)
scatter_with_gaussian_kde(axs[2], np.log10(predInterm['kcat']), predInterm['pred_log10kcat'], size = 30)
scatter_with_gaussian_kde(axs[3], np.log10(predSecond['kcat']), predSecond['pred_log10kcat'], size = 30)
for ax in axs:
    ax.set_xlim([-4.5,8])
    ax.set_ylim([-4.5,8])
    ax.set_xlabel('log10(experimental kcat 1/s)')
    ax.set_ylabel('log10(predicted kcat 1/s)')
    ax.plot([-4.5,8],[-4.5,8],'r--')
axs[0].set_title('Primary-CE(n=' +str(len(predPCE.index))+')');
axs[0].text(x=-2,y=6,s='R2='+str( get_r2(np.log10(predPCE['kcat']), predPCE['pred_log10kcat']) ))
axs[0].text(x=-2,y=5,s='RMSE='+str( get_rmse(np.log10(predPCE['kcat']), predPCE['pred_log10kcat']) ))
axs[1].set_title('Primary-AFN(n=' +str(len(predPAFN.index))+')' );
axs[1].text(x=-2,y=6,s='R2='+str( get_r2(np.log10(predPAFN['kcat']), predPAFN['pred_log10kcat']) ))
axs[1].text(x=-2,y=5,s='RMSE='+str( get_rmse(np.log10(predPAFN['kcat']), predPAFN['pred_log10kcat']) ))
axs[2].set_title('Intermediary(n=' +str(len(predInterm.index))+')');
axs[2].text(x=-2,y=6,s='R2='+str( get_r2(np.log10(predInterm['kcat']), predInterm['pred_log10kcat']) ))
axs[2].text(x=-2,y=5,s='RMSE='+str( get_rmse(np.log10(predInterm['kcat']), predInterm['pred_log10kcat']) ))
axs[3].set_title('Secondary(n=' +str(len(predSecond.index))+')');
axs[3].text(x=-2,y=6,s='R2='+str( get_r2(np.log10(predSecond['kcat']), predSecond['pred_log10kcat']) ))
axs[3].text(x=-2,y=5,s='RMSE='+str( get_rmse(np.log10(predSecond['kcat']), predSecond['pred_log10kcat']) ))
```
